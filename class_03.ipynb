Generatiing artificial datasets



```r
# A sequence from 1 to 100
id <- 1:100

# Generate numbers followig a normal distribution
ingresos <- rnorm(n = 100,mean = 2500, sd = 300)

# A manual dataframe created
ingresos_clientes <- data.frame(id, ingresos)
ingresos_clientes
```

```r
fuente_id <- 1:100
id2 <- sample(fuente_id, size = 90, replace = FALSE)
egresos <- rnorm(n = 90, mean = 500, sd = 20)
egresos_clientes <- data.frame(id2, egresos)
egresos_clientes
```

```r
# Joining dataframes
completo <- left_join(x = ingresos_clientes, y = egresos_clientes, by = c("id" = "id2")) %>% na.omit()
completo
```

how to mutate a Data Frame

```r
completo <- completo %>% mutate(ahorro = ingresos - egresos)
completo
```

Generacion **resumenes** agrupados, diferencias en los promedios de las longitudes por especies: usando tablas dinamicas.

```r
iris <- read_csv("data/iris.data")
names(iris) <- c("Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width", "Class")
iris
```

```r
# Group by Class
iris_bygroup <- iris %>% group_by(Class) %>% summarise(promedio_Sepal.Lenbgth=mean(Sepal.Length))
iris_bygroup
```
